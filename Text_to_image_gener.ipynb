```python
from diffusers import StableDiffusionPipeline
```

```python
import torch
```

```python
import requests, json
def get_completion(prompt):
  model_id = "runwayml/stable-diffusion-v1-5"
  pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
  pipe = pipe.to("cuda")
  image = pipe(prompt).images[0]
  return image
```

```python
prompt ="a dog in a park"
image = get_completion(prompt)

#the image get saved with this command
image.save("image.jpg")
```

```python
from PIL import Image
from io import BytesIO
import base64

def genrate(prompt):
  output=get_completion(prompt)
  return output

gr.close_all()
demo=gr.Interface(fn=genrate,
                  inputs=[gr.Textbox(label="Enter Prompt")],
                  outputs=[gr.Image(label="Result")],
                  title="Image Genration with Stable Diffusion",
                  description="Genration with Stable Diffusion",
                  allow_flagging="never",
                  examples=["the spirit of the tamagotch wandering in the city of Vienna","a mecha robot in a fevela"])
demo.launch(share=True, server_port=7418,debug=True)
```
